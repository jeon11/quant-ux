### Health rating by gender
Participants were asked, `How would you rate your health? (5 being healthy, 1 being not healthy)`. 

Let's breakdown the data to see how self-perception of health wellness varies by gender and different age groups. In the code below, I first quary females and males from the data. 

```python
females = data.loc[data['What is your gender?'] == 'Woman']
males = data.loc[data['What is your gender?'] == 'Man']

# columns[4] is the column for health rating
mean_males = np.mean(males[males.columns[4]])
mean_females = np.mean(females[females.columns[4]])

print("Mean of males' self-health wellness: " + str(mean_males))
print("Mean of females' self-health wellness: " + str(mean_females))
```

### Health rating by age group 
Now let's breakdown the data to see how self-perception of health wellness varies by different age groups. 

```python
age = data.groupby('What age range are you?')['How would you rate your health?'].mean()
age
```

**From the data above, we see that the age group 65 or older actually has the highest self-perception of wellness. On the other hand, youngest group (age under 18) rated the lowest.**

### Health rating by age & gender group
Now let's breakdown by both gender and different age groups to look at how the self perception of health wellness change. 

```python
age_gender = data.groupby(['What age range are you?', 'What is your gender?'])['How would you rate your health?'].mean()

age_gender
```

### Visualizing the data
Now that we have some breakdown of numbers, let's try visualizing the data.

```python
age_plot = sns.catplot(x='What age range are you?', y='How would you rate your health?', \
                       hue='What is your gender?', kind="swarm", data=data);
```

The plot above wasn't as helpful. Let's try understanding the distribution of man and woman. For simplicity sake, I only count in man and woman, and graph them using distplot and boxplot to compare the two genders distribution. 

```python
rating_males = males['How would you rate your health?']
rating_females = females['How would you rate your health?']

f, (ax_box1, ax_box2, ax_dist) = plt.subplots(3, sharex=True, 
                                              gridspec_kw= {"height_ratios": (0.3, 0.3, 1)})

# add boxplots at the top
sns.boxplot(rating_males, ax=ax_box1, color='g')
sns.boxplot(rating_females, ax=ax_box2, color='m')

ax_box1.axvline(np.mean(rating_males), color='g', linestyle='--')
ax_box2.axvline(np.mean(rating_females), color='m', linestyle='--')
plt.subplots_adjust(top=0.87)
plt.suptitle('Self health rating by gender', fontsize = 17)

sns.distplot(rating_males, ax=ax_dist, label='Man', kde=True, rug=True, color='g', norm_hist=True, bins=2)
sns.distplot(rating_females, ax=ax_dist, label='Woman', kde=True, rug=True, color='m', norm_hist=True, bins=2)

ax_dist.axvline(np.mean(rating_males), color='g', linestyle='--')
ax_dist.axvline(np.mean(rating_females), color='m', linestyle='--')

plt.legend()

plt.xlabel('Self health rating score', fontsize=16)

ax_box1.set(xlabel='')
ax_box2.set(xlabel='')
```

### Characterizing qualitative data through word cloud
Let's change focus and try analyzing textual inputs from the participants. We will analyze the column `How is your health and/or fitness information being used?` question to identify any emerging keywords using the word cloud representation. **Disclaimer: The result here is not such a useful or accurate representation as the stopwords did not clearly filter out.** 

We first call in a list of stopwords to filter out any unnecessary words, such as 'I', 'and', and etc. We then flatten out all the responses into a single list of words. 

```python
# builtin stopword sets from nltk 
stop = set(stopwords.words('english'))

# get only the valid responses (filter out NaNs)
response = [x for x in data[data.columns[-3]] if x == x]

word_dict = []
for i in range(0, len(response)):
    word_dict.append(response[i].split())
word_filtered = []

# flatten the list and lower all letter cases
for sublist in word_dict:
    for item in sublist:
        word_filtered.append(item.lower())

# remove stopwords
word_filtered = [x for x in word_filtered if x not in stop]

word_filtered = [word.replace('.','').replace(',', '').replace("'",'')  for word in word_filtered]
```

```python
Counter(word_filtered).most_common(10)
```

The 10 most common words typed by the participants were shown above. Clearly, the removing stopword process did not do much. Let's still advance to see what word cloud representation looks like. 

```python
texts = " ".join(word for word in word_filtered)
cloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(texts)
plt.imshow(cloud, interpolation='bilinear')
```

We see some useful keywords like 'monitoring', 'fitness', 'personal information', 'food intake' and so on!
